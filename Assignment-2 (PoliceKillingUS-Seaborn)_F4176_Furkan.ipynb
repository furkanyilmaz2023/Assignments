## 2: Plot 12 most dangerous cities

```python
kill["city"].value_counts().head(12)
```

```python
#DESIRED OUTPUT
```

```python
Category1 = kill["city"].value_counts().head(12).index
Numbers1 = kill["city"].value_counts().head(12).values
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation = 60)
sns.barplot(x = Category1, y = Numbers1, data=kill)
plt.title("Most Dangerous Cities", color="b");
```

```python
# with seaborn barplot

#DESIRED OUTPUT
```

```python
# with seaborn countplot

plt.figure(figsize=(9,6))
plt.xticks(rotation = 60)
sns.countplot(x = "city", data = kill, order= kill["city"].value_counts().head(12).index)
plt.title("Most Dangerous Cities", color="b");
```
